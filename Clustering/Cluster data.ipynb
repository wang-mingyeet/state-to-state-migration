```python
years = [year for year in range(2010, 2024) if year != 2020]

data = mr.migration_matrix(mr.clean_data(2010), prop = True).drop(columns = "year")
data
```

```python
states = [
    "Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", 
    "Delaware", "District of Columbia ", "Florida", "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", 
    "Kansas", "Kentucky", "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", 
    "Minnesota", "Mississippi", "Missouri", "Montana", "Nebraska", "Nevada", 
    "New Hampshire", "New Jersey", "New Mexico", "New York", "North Carolina", 
    "North Dakota", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", 
    "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", 
    "Virginia", "Washington", "West Virginia", "Wisconsin", "Wyoming"   
]
```

```python
mr.migration_matrix(mr.clean_data(2010), prop = False).loc[:,states]
```

```python
clusters_prop_time_series_pca = pd.DataFrame()
clusters_time_series_pca = pd.DataFrame()
clusters_prop_time_series = pd.DataFrame()
clusters_time_series = pd.DataFrame()

def get_data(year, prop, k, pca):
    '''
    function to return dataframe of clustering information for states over time
    args:
        - year: the year of interest
        - prop: whether we want to study proportions or raw numbers
        - k: number of clusters
        - pca: determines whether we want the pca dataframe or the individual state info
    returns:
        a dataframe
    '''

    data = mr.migration_matrix(mr.clean_data(year), prop = prop).loc[:,states]

    model = clustering.clustermodel(data, year)
    model.standardize()
    model.model_fit(k = k)

    output = pd.DataFrame()
    
    if pca:
        output = model.plot_clusters(graph = False, return_data= True)
    else:
        output = model.plot_map(graph = False, return_data= True)

    output["year"] = year
    return output

for year in years:
    clusters_prop_time_series_pca = pd.concat([clusters_prop_time_series_pca, get_data(year, prop = True, k = 4, pca = True)])
    clusters_time_series_pca = pd.concat([clusters_prop_time_series_pca, get_data(year, prop = False, k = 4, pca = True)])
    clusters_prop_time_series = pd.concat([clusters_prop_time_series, get_data(year, prop = True, k = 4, pca = False)])
    clusters_time_series = pd.concat([clusters_time_series, get_data(year, prop = False, k = 4, pca = False)])
```

```python
fig = px.scatter(
            clusters_prop_time_series_pca,
            x='PC1',
            y='PC2',
            color='cluster',
            hover_name='State',
            title='PCA Visualization of KMeans Clustering of States',
            color_discrete_sequence='Viridis',
            animation_frame='year'
            )

fig.update_traces(marker=dict(size=10, line=dict(width=1, color='DarkSlateGrey')))
fig.update_layout(title_x=0.5)
fig.show()
# use plotly to graph our clustering after PCA
```
