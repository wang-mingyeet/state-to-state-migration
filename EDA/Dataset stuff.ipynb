An important caveat: datasets from 2009 and before do not account for any foreign or US island migration. Because of this, we need to only consider domestic migration rates. This is simple for datasets 2010 and after because we have a total population value to go off of. That is the value we will use when calculating migration rates. For datasets 2009 and before, we will provide an estimate of the total population by summing over every row. However, in doing so we underestimate the population because we don't consider (nor have) information about people from foreign places. One can run stats for each state from 2010-2023 to estimate the average percentage of the state population that consist people from foreign countries and islands. Upon immediate inspection, California, Texas, Florida, and New York stand out with noticeably larger amounts of foreign migrants than other states, but we will have to run a statistical analysis on this to see which states are significantly impacted by foreign migrants, and how we should consider this as we obtain migration rates prior to 2009.

```python
for state in mig_matrix.index:
    if state in mig_matrix.columns:
        mig_matrix.at[state, state] = 0
```

```python
mig_matrix
```

```python
from sklearn.preprocessing import StandardScaler

X = mig_matrix.iloc[:,:-1].copy()
X = X.groupby(X.index).mean()
X = X.to_numpy()

scaler = StandardScaler()
X = scaler.fit_transform(X)

X
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import plotly.graph_objects as go

inertias = []
for k in range(1, 21):
    model = KMeans(n_clusters=k, random_state=0).fit(X)
    inertias.append(model.inertia_)

fig = go.Figure(data=go.Scatter(
    x=list(range(1, 21)),
    y=inertias,
    mode='lines+markers',
    text=[f"k = {k}" for k in list(range(1, 21))],
    hoverinfo='text+y'
))

fig.update_layout(
    title="Elbow Method for Optimal k",
    xaxis_title="Number of Clusters (k)",
    yaxis_title="Inertia",
    hovermode="closest"
)

fig.show()
```

From this plot it seems that $k = 4$ or $k = 5$ would be best. The inertia still is dropping but not as drastically as from before. Recall we also only have 50 points to cluster so we should not have too many clusters.

Sources:
- https://www.geeksforgeeks.org/elbow-method-for-optimal-value-of-k-in-kmeans/?
- https://scikit-learn.org/stable/modules/clustering.html#k-means

We will use 4 clusters.

```python
model = KMeans(n_clusters=4, random_state=0)
clusters = model.fit_predict(X)
```

```python
plot_df = mig_matrix.iloc[:,:-1].copy()
plot_df = plot_df.groupby(plot_df.index).mean()

#plot_df["cluster"] = clusters
plot_df
```

```python
def get_top_3_with_values(row):
    top = row.sort_values(ascending=False).head(3)
    return pd.Series([top.index[0], top.iloc[0],
                      top.index[1], top.iloc[1],
                      top.index[2], top.iloc[2]])

top_3_df = plot_df.apply(get_top_3_with_values, axis=1)
top_3_df.columns = ['Top_1', 'Prop_1', 'Top_2', 'Prop_2', 'Top_3', 'Prop_3']

top_3_df["cluster"] = clusters
top_3_df
```

```python
top_3_df['hover'] = (
    'Top 1: ' + top_3_df['Top_1'] + ' (' + (top_3_df['Prop_1']*100).round(1).astype(str) + '%)<br>' +
    'Top 2: ' + top_3_df['Top_2'] + ' (' + (top_3_df['Prop_2']*100).round(1).astype(str) + '%)<br>' +
    'Top 3: ' + top_3_df['Top_3'] + ' (' + (top_3_df['Prop_3']*100).round(1).astype(str) + '%)'
)

top_3_df['State'] = top_3_df.index.map(state_abbrev)

top_3_df
```

We will also have a clustering visualization of states after using PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['cluster'] = clusters
pca_df['State'] = plot_df.index  # or use abbreviations if you added them

pca_df
```

```python
fig = px.scatter(
    pca_df,
    x='PC1',
    y='PC2',
    color='cluster',
    hover_name='State',
    title='PCA Visualization of KMeans Clustering of States',
    color_discrete_sequence='Viridis'  # or discrete if clusters are categorical
)

fig.update_traces(marker=dict(size=10, line=dict(width=1, color='DarkSlateGrey')))
fig.update_layout(title_x=0.5)
fig.show()
```
